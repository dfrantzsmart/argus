```python
umap  = UMAP(metric='manhattan', min_dist=0.0)

umapfit = umap.fit_transform(trainDf.drop(['diagnosis'], axis=1))

umapResults = pd.DataFrame()
umapResults['diagnosis'] = trainDf['diagnosis']
umapResults['x'] = umapfit[...,0]
umapResults['y'] = umapfit[...,1]
umapResults['x'] = umapResults['x'] + umapResults['x'].max()
umapResults['y'] = umapResults['y'] + umapResults['y'].max()
```

```python
# umapResults.head()
```

```python
sns.scatterplot(x='x', y='y', size='diagnosis', data=umapResults, alpha=0.5)
```

```python
kmeans = KMeans(n_clusters=2)
kmeans = kmeans.fit(umapResults[['x', 'y']])
```

```python
# kmeans.labels_
```

```python
kmeanResults = umapResults.copy()
kmeanResults['cluster'] = kmeans.labels_
```

```python
x_train = trainSubDf.copy()
del x_train['diagnosis']
y_train = x_train.copy()[['x','y']]
x_train.drop(['x','y'], axis=1, inplace=True)
x_train = x_train.values
y_train = y_train.values
```

```python
x_train.shape
```

```python
model = kr.Sequential([
    kr.layers.Dense(240, activation=tf.nn.swish),
    kr.layers.Dense(240, activation=tf.nn.swish),
    kr.layers.Dense(240, activation=tf.nn.swish),
    kr.layers.Dense(240, activation=tf.nn.swish),
    kr.layers.Dense(240, activation=tf.nn.swish),
    kr.layers.Dense(240, activation=tf.nn.swish),
    kr.layers.Dense(240, activation=tf.nn.swish),
    kr.layers.Dense(240, activation=tf.nn.swish),
    kr.layers.Dense(240, activation=tf.nn.swish),
    kr.layers.Dense(240, activation=tf.nn.swish),
    kr.layers.Dense(240, activation=tf.nn.swish),
    kr.layers.Dense(240, activation=tf.nn.swish),
    kr.layers.Dense(240, activation=tf.nn.swish),
    kr.layers.Dense(240, activation=tf.nn.swish),
    kr.layers.Dense(240, activation=tf.nn.swish),
    kr.layers.Dense(240, activation=tf.nn.swish),
    kr.layers.Dense(240, activation=tf.nn.swish),
    kr.layers.Dense(240, activation=tf.nn.swish),
    kr.layers.Dense(240, activation=tf.nn.swish),
    kr.layers.Dense(240, activation=tf.nn.swish),
    kr.layers.Dense(240, activation=tf.nn.swish),
    kr.layers.Dense(240, activation=tf.nn.swish),
    kr.layers.Dense(240, activation=tf.nn.swish),
    kr.layers.Dense(240, activation=tf.nn.swish),
    kr.layers.Dense(240, activation=tf.nn.swish),
    kr.layers.Dense(240, activation=tf.nn.swish),
    kr.layers.Dense(240, activation=tf.nn.swish),
    kr.layers.Dense(240, activation=tf.nn.swish),
    kr.layers.Dense(240, activation=tf.nn.swish),
    kr.layers.Dense(240, activation=tf.nn.swish),
    
    kr.layers.Dense(2, activation=tf.nn.swish),
])
```

```python
# model.compile(optimizer=tf.optimizers.Adam(),
# #               loss='MeanAbsolutePercentageError',)
#               loss='mape',)
```

```python
# model.compile(optimizer=tfa.optimizers.AdamW(learning_rate=0.0001, amsgrad=True),
# #               loss='MeanAbsolutePercentageError',)
#               loss='Huber',)
```

```python
model.compile(optimizer=tf.optimizers.Adam(learning_rate=0.0001, amsgrad=True),
#               loss='MeanAbsolutePercentageError',)
              loss='Huber',)
```

```python
# model.compile(optimizer=tf.optimizers.Nadam(),
# #               loss='MeanAbsolutePercentageError',)
#               loss='mape',)
```

```python
model.fit( x_train, y_train, batch_size=2048, epochs=600)
# model.fit( x_train, y_train, epochs=600)
```

```python
trainDf, devDf, testDf = loadInitialDataSplits()
```

```python
model = kr.models.load_model(umapPredModelName)
```

```python
with open(kmeansPredModelName, 'rb') as handle:
    kmeans = pickle.load(handle)
del handle
```

```python
devDf.head()
```

```python
x_dev = devDf.copy()
devDfCopy = devDf.copy()
del x_dev['diagnosis']
x_dev = x_dev.values
x_dev.shape
```

```python
umapPreds = model.predict(x_dev)
```

```python
umapPreds.shape
```

```python
# umapPreds[:5]
```

```python
devDfCopy['xPred'] = umapPreds[...,0]
devDfCopy['yPred'] = umapPreds[...,1]
devDfCopy.shape
```

```python
devDfCopy.head()
```

```python
devDfCopy['clusterPred'] = kmeans.predict(umapPreds)
devDfCopy['kmeansPred'] = [ 0 if x == 1 else 1 for x in devDfCopy['clusterPred'].values ]
devDfCopy.shape
```

```python
# devDfCopy.head()
```

```python
sns.set(font_scale=1.5)
sns.scatterplot(x='xPred', y='yPred', hue='kmeansPred', size='diagnosis', data=devDfCopy, alpha=.5)
```

```python
le = skLabelEncoder()
le.fit(devDfCopy['diagnosis'])
```

```python
devDfCopy['diagnosis_encoded'] = le.transform(devDfCopy['diagnosis'])
devDfCopy.shape
```

```python
devDfCopy.head()
```

```python
cm = confusion_matrix(devDfCopy['diagnosis_encoded'], devDfCopy['kmeansPred'])
```

```python
sns.heatmap(cm, annot=True, 
            cmap='Blues')
plt.ylabel("True Label") 
plt.xlabel("Predicted Label") 
```
